```python
import numpy 
import matplotlib.pyplot as plt
import random
import abc
import math
x=0
difference=0
sum=0
data=[]
finall=0
z=0
time=[]
N=1000000
for n in range(1 , N+1):
     time.append(n)
     sum = 0 
     while 1:
         z=z+1
         x=random.uniform(0, 1)
         sum = x + sum
         if sum > 1 :
            finall= z + finall
            data.append(finall/n)
            z=0
            break
  
e_simulation=finall/N
print("e simulation with " + str(N) + "  test  is " + str(e_simulation))
difference=abs(e_simulation-math.e)
# plot
plt.plot(time ,data )
plt.xlabel("test time")
plt.ylabel("amount of e simulated")
plt.title('e simulation   ' + str(N)+ " test "   +  "\ne simulation finall answer = " +  str(e_simulation) + "\namount of difference between e simulated  and real e  is  " +  str(difference))
print("\namount of difference between e simulated  and real e = " +  str(difference))
plt.axhline(y = math.e, color = 'r', linestyle = '-', linewidth=0.5 , alpha=1 )
plt.show()


```
